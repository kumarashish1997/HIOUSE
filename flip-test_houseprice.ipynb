# Iqr Method

```python
q1=df1.quantile(0.25)
q1
```

```python
q2=df1.quantile(0.75)
q2
```

```python
iqr=q2-q1
iqr
```

```python
upper=q2+(iqr*1.5)
lower=q1-(iqr*1.5)
```

```python
df3=df1[(df1>lower) & (df1<upper)]
df3
```

```python
df3.info()
```
